```python
s2015["goals_Rank"] = s2015['Goals'].rank(pct=True)
s2015["passes_Rank"] = s2015['Passes'].rank(pct=True)
s2015["assits_Rank"] = s2015['Assists'].rank(pct=True)
s2015["ThrBall_Rank"] = s2015['ThrBall'].rank(pct=True)
s2015["Interception_Rank"] = s2015['Interceptions'].rank(pct=True)
s2015['Crosses_Rank'] = s2015['Crosses'].rank(pct=True)

```

```python
for i in range(len(s2015)):
    s2015['OVR'].iloc[i] = int((s2015['goals_Rank'].iloc[i]*0.1 +  s2015['assits_Rank'].iloc[i]*0.3 + s2015['passes_Rank'].iloc[i]*0.2 + s2015['Crosses_Rank'].iloc[i]*0.15+s2015['ThrBall_Rank'].iloc[i]*0.15+s2015['Interception_Rank'].iloc[i]*0.1)*100)
s2015.head()
```
