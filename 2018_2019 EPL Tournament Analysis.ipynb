## TOTAL GOAL SCORING PERFORMANCE OF ALL TEAMS PER SEASON

```python
data.head(3)
```

```python
total_goals= data.groupby(["Squad", "Season"]).agg({"Goals": "sum"}).reset_index()
s_goals= total_goals.pivot_table('Goals', ["Squad"], "Season").reset_index()
s_goals.fillna(0, inplace = True)


s_goals = s_goals.nlargest(27,"2018-2019")
```

```python
plt.rcParams["figure.figsize"] = (13,7)
seq_col = sns.color_palette("ocean", 4)
plot = s_goals[["Squad", "2015-2016", "2016-2017" ,"2017-2018", "2018-2019"]].plot(x='Squad', kind='bar', stacked = True, color = seq_col)
plot.set_title("Goal Scoring Performance of Teams per Season", fontdict = {'fontsize': 20, 'fontweight': 'bold'})
plot.set_xlabel("Teams")
plt.show()
```

## Insights

### ‣ Top 5 teams in 2018-2019 season: Mancester City, Liverpool, Arsenal, Tottenham, Manchester United
### ‣ Top 5 teams in 2017-2018 season: Mancester City, Liverpool, Tottenham, Arsenal, Manchester United
### ‣ Top 5 teams in 2016-2017 season: Tottenham, Chelsea, Liverpool, Manchester City, Arsenal
### ‣ Top 5 teams in 2015-2016 season: Mancester City, West Ham United, Leicester City, Tottenham, Arsenal

##  STRIKER ANALYSIS

### Correlation of Striker Performance Variables

```python
strikerCorrVars = Strikers[["Market Value", "Age", "Goals", "Shots", "SOT", "Assists", "App"]]
sns.pairplot(strikerCorrVars, hue = "Goals")
plt.show()
```

## Insight

### ‣ The strongest correlation with market value are goals, shots and shots on target. This indicates that before a striker is considered valuable, he must have a high shot accuracy

### Team striker goal-scoring performance


```python
strikers = Strikers[Strikers.Season == "2018-2019"]
strikers.head()
```

```python
strikers["Squad"].nunique()
```

```python
t_squadStrikers = strikers.groupby("Squad").agg({"Goals":"sum"}).nlargest(20, "Goals")
plot = sns.barplot(data = t_squadStrikers, x = "Goals", y = t_squadStrikers.index, palette= "inferno")
plot.set_title("Striker Goal Scoring Performance per Team", fontdict = {'fontsize': 20, 'fontweight': 'bold'})
plot.set_ylabel("Teams")
plt.show()
```

### Striker goal-scoring performance

```python
plot=sns.barplot(data = strikers.nlargest(15, "Goals"), x = "Goals", y = "Player", color= "g", palette= "copper_r")
plot.set_title("Top 15 Goal Scoring Strikers", fontdict = {'fontsize': 20, 'fontweight': 'bold'})
plot.set_ylabel("Striker")
plt.show()   
```

### Conversion rate

```python
top_5 = t_squadStrikers.nlargest(5, "Goals").reset_index()
top_5 = strikers.merge(top_5, on = "Squad", how = "inner")
top_5["Squad"].unique()
```

```python
top_5.shape
```

```python
top_5["Player"].nunique()
```

```python
conv_rate = ((top_5.Goals_x/top_5.SOT) * 100).round(1)
top_5["Conversion Rate (%)"] = conv_rate
plot=sns.barplot(data = top_5.nlargest(15, "Conversion Rate (%)"), y = "Player", x = "Conversion Rate (%)", color = "g")
plot.set_title("Top 15 Conversion Rates from Top 5 Teams", fontdict = {'fontsize': 20, 'fontweight': 'bold'})
plot.set_ylabel("Striker")
plt.show()   
```

### Top 3 goal scoring strikers in top 5 scoring clubs

```python
LV = top_5[top_5.Squad == "Liverpool"].nlargest(3,"Goals_x")
MC = top_5[top_5.Squad == "Manchester City"].nlargest(3,"Goals_x")
TH = top_5[top_5.Squad == 'Tottenham'].nlargest(3,"Goals_x")
CH = top_5[top_5.Squad == "Chelsea"].nlargest(3,"Goals_x")
Ars = top_5[top_5.Squad == "Arsenal"].nlargest(3,"Goals_x")
```

```python
top5_3Strikers = pd.concat([LV, MC, TH, CH, Ars], ignore_index=True)
plot = sns.barplot(x = "Player", y = "Goals_x", hue= "Squad", data = top5_3Strikers, palette= "gist_earth")
plt.xticks(rotation=90)
plot.set_title("Top 3 Goal Scoring Strikers from Top 5 Goal Scoring Teams", fontdict = {'fontsize': 20, 'fontweight': 'bold'})
plot.set_xlabel("Striker")
plot.set_ylabel("Goals")
plt.show()
```

```python
top_5 = defend.groupby("Squad").agg({"Clears":"sum"}).nlargest(5, "Clears").reset_index()
top_5 = defend.merge(top_5, on = "Squad", how = "inner")
top_5["Squad"].unique()
```

```python
TH = top_5[top_5.Squad == "Fulham"].nlargest(3,"Clears_x")
SN = top_5[top_5.Squad == "Cardiff City"].nlargest(3,"Clears_x")
EV = top_5[top_5.Squad == "Burnley"].nlargest(3,"Clears_x")
LC = top_5[top_5.Squad == "Southampton"].nlargest(3,"Clears_x")
WB = top_5[top_5.Squad == "Newcastle United"].nlargest(3,"Clears_x")

top5_3Defenders = pd.concat([TH, SN, EV, LC, WB], ignore_index=True)
plot = sns.barplot(x = "Player", y = "Clears_x", hue= "Squad", data = top5_3Defenders, palette= "Dark2_r")
plt.xticks(rotation=90)
plot.set_title("Top 3 Defenders for Clearance from Top 5 Teams with Clearances", fontdict = {'fontsize': 20, 'fontweight': 'bold'})
plot.set_xlabel("Defender")
plot.set_ylabel("Clears")
plt.show()
```
